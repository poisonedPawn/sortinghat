```python
X = []
y = []
for img in glob.glob(path+"*"):
    house = img.split("\\")[-1].split("_")[0]
    im = cv.imread(img,cv.IMREAD_GRAYSCALE)
    im = im.flatten()
    X.append(im)
    y.append(house)
    

```

```python
y
```

```python
df_X = pd.DataFrame(X)
df_y = pd.DataFrame(y, columns= ["house"])
pics = pd.concat([df_X,df_y], axis=1)
```
